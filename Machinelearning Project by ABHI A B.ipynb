## FEATURE ENGINEERING

```python
# outlier detection and its correction

numerical_cols = ['Age', 'Flight Distance', 'Departure Delay in Minutes', 'Arrival Delay in Minutes']
for i in numerical_cols:
    plt.figure(figsize=(10, 5))
    sns.boxplot(y=df[i], color='skyblue')
    plt.title("Boxplot of " + i)
    plt.show()
```

```python
for i in numerical_cols:
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR


    outliers = df[(df[i] < lower_bound) | (df[i] > upper_bound)]
    print("\nColumn:", i)
    print("IQR:", IQR)
    print("Lower Bound:", lower_bound)
    print("Upper Bound:", upper_bound)



    df[i] = df[i].clip(lower=lower_bound, upper=upper_bound)
```

## MODEL EVALUATION

```python
#Model Selection and Hyperparameter Grid Setup

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier , GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models = {
    'Logistic Regression': LogisticRegression(max_iter=1000),
    'Decision Tree Classifier': DecisionTreeClassifier(),
    'Random Forest Classifier': RandomForestClassifier(),
    'Gradient Boosting Classifier': GradientBoostingClassifier(),
    'SVC': SVC(),
    'KNN Classifier': KNeighborsClassifier()
}

param_grid = {
    'Logistic Regression': {
        'penalty': ['l2'],
        'C': [0.1, 1, 10],
        'solver': ['lbfgs']
    },
    'Decision Tree Classifier': {
        'max_depth': [None, 5, 10, 20],
        'min_samples_split': [2, 5, 10]
    },
    'Random Forest Classifier': {
        'n_estimators': [50, 100],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5]
    },
    'Gradient Boosting Classifier': {
        'n_estimators': [50, 100],
        'learning_rate': [0.1, 0.05],
        'max_depth': [3, 5]
    },
    'SVC': {
        'C': [0.1, 1, 5],
        'kernel': ['linear', 'rbf'],
        'gamma': ['scale', 'auto']
    },
    'KNN Classifier': {
        'n_neighbors': [3, 5, 7],
        'weights': ['uniform', 'distance']
    }
}

```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score,classification_report
```

```python
# hyperparameter tuning using GridSearchCV for multiple models and evaluating them based on accuracy and classification metrics.


result = []

for name in models:
    print(f'Tuning: {name}')
    grid = GridSearchCV(models[name], param_grid[name], cv=3, scoring='accuracy', n_jobs=-1)
    grid.fit(xtrain, ytrain)

    best_model = grid.best_estimator_
    y_pred = best_model.predict(xtest)
    print(f'{grid.best_params_}')
    acc = accuracy_score(ytest, y_pred)
    print(classification_report(ytest, y_pred))  # Optional detailed metrics

    result.append({
        'Model': name,
        'Best Params': grid.best_params_,
        'Test Accuracy': acc
    })
```

```python
results_df = pd.DataFrame(result)
results_df
```

```python
best_model_idx = results_df['Test Accuracy'].idxmax()
worst_model_idx = results_df['Test Accuracy'].idxmin()
```

```python
print("Best Performing Model:", results_df.loc[best_model_idx, 'Model'])
print("Worst Performing Model:", results_df.loc[worst_model_idx, 'Model'])
```

```python
#Training Random Forest with Selected Hyperparameters and Evaluating Accuracy

m = RandomForestClassifier(max_depth=None, min_samples_split=5, n_estimators=50)
m.fit(xtrain, ytrain)
y_pred = m.predict(xtest)


print("Accuracy:", accuracy_score(ytest, y_pred))
```
